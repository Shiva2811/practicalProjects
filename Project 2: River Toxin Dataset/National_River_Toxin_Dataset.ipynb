```python
#importing required libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#load dataset
data = pd.read_csv('National_River_Toxin_Dataset.csv')
data.head(10)
```

```python
# Check basic information
data.info()
```

```python
# Check for missing values
print("\nMissing Values:")
data.isnull().sum()
```

```python
# Handle missing values (Example: Filling with mean for numeric columns)
# Fill numeric columns with mean
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())

# Fill non-numeric columns with a placeholder or mode
non_numeric_columns = data.select_dtypes(exclude=['float64', 'int64']).columns
data[non_numeric_columns] = data[non_numeric_columns].fillna("Unknown")
# Fill missing numeric values with column mean
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
numeric_columns = numeric_columns.fillna(numeric_columns.mean())
```

```python
# Check for duplicates
print("\nChecking for duplicates...")
print(f"Number of duplicate rows: {data.duplicated().sum()}")
data = data.drop_duplicates()
```

```python
# Descriptive statistics
print("\nDescriptive Statistics:")
data.describe()
```

```python
# Correlation matrix
# Select only numeric columns for correlation calculation
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_columns.corr()

# Print and visualize the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Distribution of pH levels
plt.figure(figsize=(8, 5))
sns.histplot(data['pH_Level'], kde=True, bins=20, color='blue')
plt.title("Distribution of pH Levels")
plt.xlabel("pH Level")
plt.ylabel("Frequency")
plt.show()
```

```python
# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Filter data for a specific date range
start_date = '2020-01-01'
end_date = '2023-12-31'
filtered_data = data[(data['Date'] >= start_date) & (data['Date'] <= end_date)]

# Line plot: Toxin levels over time for each river
plt.figure(figsize=(12, 6))
for river in data['River_System'].unique():
    subset = data[data['River_System'] == river]
    plt.plot(subset['Date'], subset['Lead'], label=f"{river} - Lead")
plt.xticks(rotation=45)
plt.title("Lead Levels Over Time by River System")
plt.xlabel("Date")
plt.ylabel("Lead Level (mg/L)")
plt.legend()
plt.show()
```

```python
#Boxplot: Toxin levels by rever system
plt.figure(figsize=(10,6))
sns.boxplot(x='River_System', y='Lead', data =data)
plt.title('Lead levels by rever system')
plt.xlabel('Rever System')
plt.ylabel('Lead Level(mg/L)')
plt.show()
```

```python
#Aggregated insights: Average toxin levels per river system
average_toxins = data.groupby('River_System')[['Lead','Mercury','Arsenic','Nitrates','Phosphates']].mean()
print("\nAverage Toxin Levels by River System:")
print(average_toxins)
```

```python
#Bar chat for average toxin levels
average_toxins.plot(kind='bar', figsize=(10,6))
plt.title("Average Toxin Levels by River System")
plt.ylabel("Concentration (mg/L)")
plt.xlabel("River System")
plt.xticks(rotation=45)
plt.legend(title="Toxin")
plt.show()
```

```python
# Dissolved Oxygen levels over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Dissolved_Oxygen', hue='River_System', data=data)
plt.title("Dissolved Oxygen Levels Over Time")
plt.xlabel("Date")
plt.ylabel("Dissolved Oxygen (mg/L)")
plt.xticks(rotation=45)
plt.legend(title="River System")
plt.show()
```
