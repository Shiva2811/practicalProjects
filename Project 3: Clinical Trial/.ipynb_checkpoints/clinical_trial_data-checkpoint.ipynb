```python
#import reqiure libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
#Loading dataset
data = pd.read_csv('diabetes_medication_clinical_trial1.csv')
data.head(10)
```

```python
# Check data structure and types
print("\nDataset Info:")
data.info()
```

```python
#Check for missing values
print('\nMissing values')
data.isnull().sum()
```

```python
#Handle missing values(example fill or drop)
#Filling missing Adverse_events with 'None' and numeric column wiht their mean
data['adverse_event'].fillna('None', inplace=True)
numeric_column = data.select_dtypes(include=['float64','int64']).columns
data[numeric_column] = data[numeric_column].fillna(data[numeric_column].mean())
```

```python
#Verify no missing value remains
print("\nMissing Values After Cleaning:")
print(data.isnull().sum())
```

```python
# Descriptive statistics
print("\nDescriptive Statistics:")
print(data.describe())
```

```python
# Group comparisons: Baseline HbA1c levels
group_summary = data.groupby('group')[['initial_hba1c', 'age', 'duration_diabetes']].mean()
print("\nGroup Summary (Mean Values):")
print(group_summary)
```

```python
# Statistical test: Compare baseline HbA1c levels between groups
t_stat, p_value = ttest_ind(
    data[data['group'] == 'treatment']['initial_hba1c'],
    data[data['group'] == 'placebo']['initial_hba1c']
)
print(f"\nT-Test Results: t-stat = {t_stat:.2f}, p-value = {p_value:.4f}")
```

```python
# Visualizations
# Boxplot of initial HbA1c by group
plt.figure(figsize=(8, 6))
sns.boxplot(x='group', y='initial_hba1c', data=data, hue='gender', palette='Set2')
plt.title("Baseline HbA1c Levels by Group and Gender")
plt.xlabel("Group")
plt.ylabel("Initial HbA1c")
plt.legend(title="Gender")
plt.show()
```

```python
# Adverse Events Frequency
plt.figure(figsize=(8, 6))
sns.countplot(y='adverse_event', hue='group', data=data, palette='Set3')
plt.title("Adverse Events by Group")
plt.xlabel("Count")
plt.ylabel("Adverse Event")
plt.legend(title="Group")
plt.show()
```

```python
# HbA1c levels over time by group
plt.figure(figsize=(10, 6))
sns.lineplot(x='time', y='hba1c', hue='group', data=data, marker='o', palette='Set1')
plt.title("HbA1c Levels Over Time by Group")
plt.xlabel("Time (Months)")
plt.ylabel("HbA1c Level")
plt.legend(title="Group")
plt.show()
```

```python
# Age distribution by group
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='age', hue='group', kde=True, palette='Set2', bins=20)
plt.title("Age Distribution by Group")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# HbA1c improvement (initial - final)
data['hba1c_change'] = data['initial_hba1c'] - data['hba1c']
change_summary = data.groupby('group')['hba1c_change'].mean()
print("\nAverage HbA1c Change by Group:")
print(change_summary)
```

```python
# Bar plot for HbA1c change
plt.figure(figsize=(8, 6))
sns.barplot(x='group', y='hba1c_change', data=data, hue='gender',palette='Set2')
plt.title("Average HbA1c Change by Group")
plt.xlabel("Group")
plt.ylabel("Average Change in HbA1c")
plt.show()
```
